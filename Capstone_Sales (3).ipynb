# Outlier treatment



```python
df.Sales.shape
```

```python
Q1 = df.Sales.quantile(0.25)
Q3 = df.Sales.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df.Sales < (Q1 - 1.5 * IQR)) |(df.Sales > (Q3 + 1.5 * IQR)))]
df.Sales.shape
```

```python
# we can see some outliers in the Sales column but it can be nullified
```
